```python
#QUESTION1------

A for loop and a while loop are two common types of loop constructs in programming. 
The main difference between the two is that a for loop is used when you know the number of iterations beforehand, 
whereas a while loop is used when you don't know the number of iterations beforehand.

Here's an example of when to use a for loop:

# Print the numbers from 1 to 10
for i in range(1, 11):
    print(i)
	
In this case, we know that we want to print the numbers from 1 to 10, so we use a for loop to iterate over the range of numbers. 
The range() function generates a sequence of numbers from 1 to 10, and the for loop allows us to execute the print() statement 10 times.

Here's an example of when to use a while loop:

# Keep asking the user for input until they enter "stop"

user_input = ""
while user_input != "stop":
    user_input = input("Enter something: ")
	
In this case, we don't know how many times we'll have to ask the user for input, so we use a while loop.
 The loop continues to execute as long as the user_input variable does not equal "stop".
 When the user finally enters "stop", the condition in the while loop becomes false, and the loop terminates.
 
```

```python
#QUESTION3-----

units_consumed = int(input('Enter units consumed in a month: '))
total_bill = 0

if units_consumed<=100:
    total_bill = units_consumed*4.5
elif units_consumed>100 and units_consumed <=200:
    greater_than_hun_units = units_consumed-100
    total_bill = ((greater_than_hun_units*6)+(100*4.5))
elif units_consumed>200 and units_consumed <=300:
    greater_than_twohun_units = units_consumed-200
    total_bill = ((100*6)+(greater_than_twohun_units*10)+(100*4.5))
elif units_consumed>300:
    greater_than_threehun_units = units_consumed-300
    total_bill = ((100*6)+(100*10)+(greater_than_threehun_units*20)+(100*4.5))
    
print('Total Bill: ', total_bill)
	
```

```python
#QUESTION4--------

result_list = []

for num in range(1, 101):
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        result_list.append(num)

print(result_list)
```
