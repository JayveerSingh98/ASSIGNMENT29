```python
#Q1
In object-oriented programming (OOP), a class is a blueprint or template for creating objects. 
It defines a set of properties and methods that the objects of the class will possess. 
An object is an instance of a class, created from the blueprint provided by the class. 
It has its own set of properties and methods, and can interact with other objects and classes in the program.

Let's take an example of a class and object in the context of a car.
A Car class can be defined with properties such as make, model, color, and year, and methods 
such as start(), accelerate(), and stop(). These properties and methods define the characteristics and behavior of a car object.

Here's an example:

class Car:
    def __init__(self, make, model, color, year):
        self.make = make
        self.model = model
        self.color = color
        self.year = year
        self.speed = 0

    def start(self):
        print("The car has started.")

    def accelerate(self):
        self.speed += 10
        print(f"The car is now traveling at {self.speed} mph.")

    def stop(self):
        self.speed = 0
        print("The car has stopped.")

my_car = Car("Toyota", "Corolla", "blue", 2022)

In this example, we have defined a Car class with four properties (make, model, color, year) and
three methods (start, accelerate, stop). The __init__ method is a special method that is called
when a new object is created from the class, and it initializes the object's properties.

We then create an object of the Car class named my_car with the values "Toyota", "Corolla", "blue",
and 2022 passed as arguments to the __init__ method. This object has its own set of properties and methods,
separate from any other Car objects created from the same class.

```

```python
#Q1
class Car:
    def __init__(self, make, model, color, year):
        self.make = make
        self.model = model
        self.color = color
        self.year = year
        self.speed = 0

    def start(self):
        print("The car has started.")

    def accelerate(self):
        self.speed += 50
        print(f"The car is now traveling at {self.speed} mph.")

    def stop(self):
        self.speed = 0
        print("The car has stopped.")

my_car = Car("Toyota", "Corolla", "blue", 2022)
```

```python
#Q3
In Python, __init__() is a special method that is called when an object is created from a class. 
It is used to initialize the object's properties with default or user-defined values.
The __init__() method is called a constructor method, as it constructs the object and sets its initial values.

Here's an example :

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"My name is {self.name} and I am {self.age} years old.")

person1 = Person("Alice", 25)
person2 = Person("Bob", 30)

person1.introduce()
person2.introduce()

In this example, we have a Person class with two properties (name and age) and a method (introduce) that prints out the
person's name and age. The __init__() method takes two arguments - name and age - and sets the name and age properties of
the object being created with those values.

We then create two Person objects named person1 and person2, passing in different values for name and age.
When the __init__() method is called for each object, it sets their respective name and age properties.

Finally, we call the introduce() method for each object, which prints out their name and age using the properties
set by the __init__() method.
```

```python
#Q3
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"My name is {self.name} and I am {self.age} years old.")

person1 = Person("Alice", 25)
person2 = Person("Bob", 30)

person1.introduce()
person2.introduce()
```

```python
#Q4
In object-oriented programming (OOP), self is a reference to the current object being manipulated. 
It is used to access the properties and methods of the object, as well as to call other methods on the object.

When a method is called on an object, the object itself is passed as the first argument to the method,
which is why the first parameter of a method is typically named self. By convention, the self parameter 
is used to refer to the current object being manipulated within the method.

Here's an example:


class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

rect = Rectangle(5, 10)
print(rect.area())
print(rect.perimeter())

In this example, we have a Rectangle class with a __init__() method that sets the width and height properties of the object, and two methods
(area() and perimeter()) that use the self parameter to access these properties and perform calculations.

When we create a new Rectangle object named rect with the values 5 and 10, the __init__() method is called with rect as the self parameter, 
setting the width and height properties of the rect object.

We can then call the area() and perimeter() methods on the rect object, which use the self parameter to access its width and height properties
and perform calculations on them. By using self, we can ensure that each method operates on the correct object and its associated properties.
```

```python
#Q4
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

rect = Rectangle(5, 10)
print(rect.area())
print(rect.perimeter())
```
